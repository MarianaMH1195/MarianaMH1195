```python
text_columns = ['first_name', 'last_name', 'email', 'address', 'district', 'city', 'country']
for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
df['rental_date'] = pd.to_datetime(df['rental_date'])
df['return_date'] = pd.to_datetime(df['return_date'])
df['payment_date'] = pd.to_datetime(df['payment_date'])
```

```python
df['amount'].describe()
df['rental_duration'].describe()
```

```python
df = df[df['amount'] > 0]
df = df[df['rental_duration'] > 0]
```

```python

df['is_active'] = df['active'].apply(lambda x: True if x==1 else False)


df['rental_month'] = df['rental_date'].dt.month
```

```python
df.to_csv('customer_activity_clean.csv', index=False)
```
